```python
import csv

# Function to extract and print employee names from a CSV file
def extract_employee_names(input_file):
    try:
        with open(input_file, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                employee_name = row.get('name', '')  # Assuming 'name' is the column name
                if employee_name:
                    print(employee_name)
    except FileNotFoundError:
        print("File not found. Please check the file path.")

if __name__ == "__main__":
    input_file = input("Enter the path to the input CSV file: ")
    extract_employee_names(input_file)
```

```python
# Function to read data from a CSV file with different headers
def read_csv_file(input_file):
    try:
        with open(input_file, 'r') as file:
            reader = csv.DictReader(file)
            records = [(row['Employee Name'], row['Date']) for row in reader]
            return records
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return []
```

```python
import csv

# Function to check for employees who worked for a specified number of consecutive days
def check_consecutive_days(records, days):
    consecutive_days_employees = []
    current_employee = None
    consecutive_days = 0

    for record in records:
        name, date = record
        if name != current_employee or not current_employee:
            consecutive_days = 1
            current_employee = name
        else:
            consecutive_days += 1

        if consecutive_days == days:
            consecutive_days_employees.append(current_employee)

    return consecutive_days_employees

# Function to read data from a CSV file
def read_csv_file(input_file):
    try:
        with open(input_file, 'r') as file:
            reader = csv.DictReader(file)
            records = [(row['Employee Name'], row['Pay Cycle Start Date']) for row in reader]
            return records
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return []

if __name__ == "__main__":
    input_file = input("Enter the path to the input CSV file: ")
    records = read_csv_file(input_file)

    if records:
        # Check for employees who worked for 7 consecutive days
        consecutive_days_employees = check_consecutive_days(records, 7)
        print("Employees who worked for 7 consecutive days:")
        for employee in consecutive_days_employees:
            print(employee)
```

```python
from datetime import datetime

def check_shift_timings(records, min_hours, max_hours):
    shift_timings_employees = []
    employee_shifts = {}
    for record in records:
        name, Time = record
        if Time:
            Time = datetime.strptime(Time, '%m/%d/%Y %I:%M %p')

            if name in employee_shifts:
                previous_shift_end = employee_shifts[name]
                time_between_shifts = (Time - previous_shift_end).total_seconds() / 3600

                if min_hours < time_between_shifts < max_hours:
                    shift_timings_employees.append(name)

            employee_shifts[name] = Time

    return shift_timings_employees


```

```python
# Function to check for employees who worked for more than 14 hours in a single shift
def check_long_shifts(records, max_hours):
    long_shift_employees = []
    for record in records:
        name, Time_str, Timeout_str = record

        # Use different variable names for parsing the date-time strings
        try:
            Time = datetime.strptime(Time_str, '%Y-%m-%d %H:%M:%S')
            Timeout = datetime.strptime(Timeout_str, '%Y-%m-%d %H:%M:%S')

            shift_duration = (Timeout - Time).total_seconds() / 3600  # Calculate shift duration in hours

            if shift_duration > max_hours:
                long_shift_employees.append(name)
        except ValueError:
            print(f"Skipping record with invalid date-time format: {record}")

    return long_shift_employees
```

```python
# Function to read data from a CSV file
def read_csv_file(input_file):
    try:
        with open(input_file, 'r') as file:
            reader = csv.DictReader(file)
            records = [(row['Employee Name'], row['Time'], row['Timecard Hours (as Time)']) for row in reader]
            return records
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return []
```

```python
if __name__ == "__main__":
    input_file = input("Enter the path to the input CSV file: ")
    records = read_csv_file(input_file)

    if records:
        # Check for employees who worked for more than 14 hours in a single shift
        max_hours = 14  # Adjust this value as needed
        long_shift_employees = check_long_shifts(records, max_hours)
        print("\nEmployees who worked for more than 14 hours in a single shift:")
        for employee in long_shift_employees:
            print(employee)
```
